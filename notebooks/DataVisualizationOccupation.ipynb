```python
import param
import pandas as pd
import panel as pn
import hvplot.pandas

pn.extension()

class Dashboard(param.Parameterized):

    def __init__(self, **params):
        self.intro_d19 = self.get_Introdata
        self.deep_d19 = self.get_deepdata
        
        self.top_widget = pn.widgets.IntSlider(name='top',start =0,end =len(self.intro_d19),step =1)
        self.feature_widget = pn.widgets.Select(name='feature',options = list(self.intro_d19.iloc[:,2:].columns),value =list(self.intro_d19.iloc[:,2:].columns))
        
        self.dtop_widget = pn.widgets.IntSlider(name='dtop',start =0,end =len(self.deep_d19),step =1)
        self.dfeature_widget = pn.widgets.Select(name='dfeature',options = list(self.deep_d19.iloc[:,6:].columns),value =list(self.deep_d19.iloc[:,6:].columns))
        mask = self.deep_d19[self.deep_d19['Occupation type']=='Summary']
        self.doccupation_widget = pn.widgets.Select(name='doccupation',options = list(mask['2019 National Employment Matrix title']), value = list(mask['2019 National Employment Matrix title']))
        
        super().__init__(**params)

        self.layout1 = ('Introduction and EDA',
             pn.Row(
                 self.create_intro_tab_text,
                 pn.Column(
                 ),
             ),
            )
        self.layout2 = ('Deep plots',
            pn.Column(
                pn.WidgetBox(
                     self.dtop_widget,
                     self.dfeature_widget,
                     self.doccupation_widget,
                 ),
                 pn.Row(self.create_layout2_bar_plot,self.get_layout2_labels_bar)
             )
            )
        self.layout3 = ('Intro plots',
             pn.Column(
                 pn.WidgetBox(
                     self.top_widget,
                     self.feature_widget,
                 ),
                 pn.Row(self.create_layout3_bar_plot,self.get_layout3_labels_bar)
             )
            )
        self.layout = pn.Tabs(self.layout1, self.layout3, self.layout2)

    @property
    def get_Introdata(self):
        d19 = pd.read_excel(occupation19, sheet_name=1)
        d19.columns = d19.iloc[0]  
        d19 = d19[1:]
        d19 = d19.dropna(thresh=len(d19.columns) - 3)
        cols = d19.columns
        d19[cols[2:]] = d19[cols[2:]].apply(pd.to_numeric, errors='coerce')
        return d19


    @property
    def get_deepdata(self):
        df = pd.read_excel(occupation19, sheet_name=2)
        #adjust column
        df.columns = df.iloc[0]
        df = df[1:]
        df = df.dropna(thresh=len(df.columns) - 3)
        df[df.columns[3:]] = df[df.columns[3:]].apply(pd.to_numeric, errors='coerce')
        df = df.drop_duplicates(subset="2019 National Employment Matrix title", keep="first")

        df.insert(3,'4 digit Matrix code',df['2019 National Employment Matrix code'].str[:4])
        df.insert(3,'Matrix code',df['2019 National Employment Matrix code'].str[:2])
        df["Matrix code"] = df["Matrix code"].apply(lambda x: f"{x}-0000")
        
        d19 = self.intro_d19[self.intro_d19.columns[:2]]
        d19.columns = ['title','Matrix code']
        
        df = pd.merge(df, d19, on='Matrix code', how='left')
        df.insert(0, 'title', df.pop('title'))
        df.insert(0, 'Matrix code', df.pop('Matrix code'))
        # print(df.info())
        # print(df.head())
        return df
 
    @param.depends("top_widget.value","feature_widget.value",)
    def create_layout3_bar_plot(self):
        data = self.intro_d19
        data.sort_values(by =self.feature_widget.value,inplace=True,ascending=False)
        data = data.head(self.top_widget.value)
        return data.hvplot.bar(data.columns[1], self.feature_widget.value,height=600, width=800)

    @param.depends("top_widget.value","feature_widget.value",)
    def get_layout3_labels_bar(self):
        data = self.intro_d19
        data.sort_values(by =self.feature_widget.value,inplace=True,ascending=False)
        data = data[[data.columns[1],data.columns[0]]]
        data = data.head(self.top_widget.value)
        data.set_index(data.columns[1],inplace=True)
        return pn.widgets.DataFrame(data, name='DataFrame')

    @param.depends("dtop_widget.value","dfeature_widget.value","doccupation_widget.value",)
    def create_layout2_bar_plot(self):
        data = self.deep_d19
        #occ4 = data[data['2019 National Employment Matrix title'] == self.doccupation_widget.value][['4 digit Matrix code']]
        #occm = data.loc[data['2019 National Employment Matrix title'] == self.doccupation_widget.value, 'Matrix code']
        #print(occ4)
        #print(occm)
        # if occ4[-1] == '0':
        #     data = data[data['Matrix code']==occm]
        # else:
        #     data = data[data['4 digit Matrix code']==occ4]
        data.sort_values(by =self.dfeature_widget.value,inplace=True,ascending=False)
        data = data.head(self.dtop_widget.value)
        return data.hvplot.bar(data.columns[3], self.dfeature_widget.value,height=600, width=800)
        
    @param.depends("dtop_widget.value","dfeature_widget.value","doccupation_widget.value",)
    def get_layout2_labels_bar(self):
        data = self.deep_d19
        #occ4 = data.loc[data['2019 National Employment Matrix title'] == self.doccupation_widget.value, '4 digit Matrix code']
        # occm = data.loc[data['2019 National Employment Matrix title'] == self.doccupation_widget.value, 'Matrix code']
        # if occ4[-1] == '0':
        #     data = data[data['Matrix code']==occm]
        # else:
        #     data = data[data['4 digit Matrix code']==occ4]
        data.sort_values(by =self.dfeature_widget.value,inplace=True,ascending=False)
        data = data[[data.columns[3],data.columns[4]]]
        data = data.head(self.dtop_widget.value)
        data.set_index(data.columns[1],inplace=True)
        return pn.widgets.DataFrame(data, name='DataFrame')
        
    @property
    def create_intro_tab_text(self):
        text = """ 
        """
        title = "# Introduction and EDA"
        return pn.Column(
            title, pn.layout.Divider(), text,
        )

d = Dashboard()
d.layout
```
